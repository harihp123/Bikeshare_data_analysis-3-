```python
import pandas as pd
import numpy as np
import datetime
import time
```

```python
city= input("welcome! which city data do you want to see? ")
if city.upper()=='CHICAGO':
         city_file= pd.read_csv('chicago.csv')
elif city.upper()=='WASHINGTON':
         city_file=pd.read_csv('washington.csv')
elif city.upper()=='NEW YORK CITY':
     city_file=pd.read_csv('new_york_city.csv')
        
print("Coverting start time and End time to timestamps:")
        

def next_statistic():
    print("\t\t...Calcluating the next statistic...")
    

city_file['Start Time']=pd.to_datetime(city_file['Start Time'])
city_file['End Time']=pd.to_datetime(city_file['End Time'])

city_file['Start Month']=city_file['Start Time'].dt.month
city_file['Start Day']=city_file['Start Time'].dt.day
city_file['Weekday']=city_file['Start Time'].dt.weekday
city_file['Difference']=city_file['End Time']-city_file['Start Time']
city_file['Start Hour']=city_file['Start Time'].dt.hour


def raw_data():
    print("The first five rows of the data are\n",city_file.head())
    print("The last five rows of the data are\n",city_file.tail())
    

raw=input("Would you like to see some raw data?type Yes or No")
if(raw.upper()=='YES'):
    raw_data()
elif(raw.upper()=='NO'):
    print("Continuing!")


print("Before you start filtering the data, here are some statistics about the enitre dataset:")

#most popular month
start_time=time.time()
print("The most popular month for the city you chose is:",city_file['Start Month'].value_counts().idxmax())
print("Time taken to calclulate this statistic:",time.time()-start_time)
next_statistic()

#most popular day
start_time=time.time()
print("The most popular day for the city you chose is:",city_file['Start Day'].value_counts().idxmax())
print("Time taken to calculate the statistic:",time.time()-start_time)
next_statistic()



#Filtering data

filtering=input("Would you like to filter the data by month,day or none")
if(filtering.upper()=='MONTH'):
    month=int(input("Enter the integer number month by which you would like to filter the data(1-6)"))
    city_file=city_file[city_file['Start Month']==month]
elif(filtering.upper()=='DAY'):
    day=int(input("Enter the day by which you would like to filter the data"))
    city_file=city_file[city_file['Start Day']==day]
else:
    print("Continuing!")

#oldest person in the data

if(city.upper()=='CHICAGO' or city.upper()=='NEW YORK CITY'):
    start_time=time.time()
    print("The oldest person was born in the year:",city_file['Birth Year'].min())
    print("The time taken to calculate the statistc is:",time.time()-start_time)
    next_statistic()
    
elif(city.upper()=='WASHINGTON'):
    print("The oldest person is not available in the dataset for washington!")
    
#youngest person in the data

if(city.upper()=='CHICAGO' or city.upper()=='NEW YORK CITY'):
    start_time=time.time()
    print("The oldest person was born in the year:",city_file['Birth Year'].max())
    print("The time taken to calculate the statistc is:",time.time()-start_time)
    next_statistic()
    
elif(city.upper()=='WASHINGTON'):
    print("The youngest person is not available in the dataset for washington!")
    
#most common birth year
if(city.upper()=='CHICAGO' or city.upper()=='NEW YORK CITY'):
    start_time=time.time()
    print("The most common year of birth is:",city_file['Birth Year'].value_counts().idxmax())
    print("The time taken to calculate the statistc is:",time.time()-start_time)
    next_statistic()
    
elif(city.upper()=='WASHINGTON'):
    print("The year of birth is not available in the dataset for washington!")


# number of user types

start_time=time.time()
print("The different types of users and the number of that particular user are:",city_file['User Type'].value_counts())
print("Time taken to calulate that statistic:",time.time()-start_time)
next_statistic()


#counts for geneder
start_time=time.time()
if(city.upper()=='CHICAGO' or city.upper()=='NEW YORK CITY'):
    print("The gender count is :",city_file['Gender'].value_counts())
    print("The time taken to calculate the statistic is:",time.time()-start_time)
    next_statistic()
elif(city.upper()=='WASHINGTON'):
    print("The gender column is unavailble for washington!")

    
# total trip duration
start_time=time.time()
print("the sum of total trip duration is:",city_file['Difference'].sum())
print("The time taken to calculate the statistc is:",time.time()-start_time)
next_statistic()

#Average trip duration

start_time=time.time()
print("the average of total trip duration is:",city_file['Difference'].mean())
print("The time taken to calculate the statistc is:",time.time()-start_time)
next_statistic()

# most common start station is
start_time=time.time()
print("The most frequent start station is:",city_file.groupby('Start Station')['Start Station'].count().idxmax())
print("The time taken to calculate the statistc is:",time.time()-start_time)
next_statistic()
    
# most common trip
start_time=time.time()
print("The most common trip is:",city_file.groupby(['Start Station','End Station']).size().idxmax())
print("The time taken to calculate the statistc is:",time.time()-start_time)
next_statistic()

# most common end station
start_time=time.time()
print("The most frequent end station is:",city_file.groupby('End Station')['End Station'].count().idxmax())
print("The time taken to calculate the statistc is:",time.time()-start_time)
next_statistic()

# most frequent hour in start time is
start_time=time.time()
print("The follwing hour occours the most frquently in the city you have chosen:",city_file['Start Hour'].value_counts().idxmax())
print("The time taken to calculate the statistc is:",time.time()-start_time)
next_statistic()


#The most frequent weekday
start_time=time.time()
print("The most frequently occouring weekday is:",city_file['Weekday'].value_counts().idxmax())


raw_data_1=input("Would you like to see some raw data once again?Type Yes or No")

if(raw_data_1.upper()=='YES'):
    raw_data()
else:
    print("Thank you for using the system!")




        
```
